```python
import os
import pydicom
import numpy as np
import cv2
from tqdm import tqdm

# Load DICOM
def load_dicom(path):
    ds = pydicom.dcmread(path)
    img = ds.pixel_array.astype(np.float32)

    # Fix MONOCHROME1 inversion
    if ds.PhotometricInterpretation == "MONOCHROME1":
        img = np.max(img) - img

    return img


# Convert to grayscale
def to_gray(img):
    return img if len(img.shape) == 2 else cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)


# Resize first (as per your chosen pipeline)
def resize_image(img, size=1024):
    return cv2.resize(img, (size, size), interpolation=cv2.INTER_AREA)


# Min-max normalization
def minmax_normalize(img):
    img = (img - img.min()) / (img.max() - img.min() + 1e-8)
    return (img * 255).astype(np.uint8)


# CLAHE enhancement
def apply_clahe(img):
    return cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8,8)).apply(img)


# Denoising
def denoise(img):
    img = cv2.medianBlur(img, 3)
    img = cv2.GaussianBlur(img, (3,3), 0)
    return img


# Crop black borders
def crop_borders(img):
    mask = img > 5
    coords = np.argwhere(mask)
    
    if coords.size == 0:
        return img
    
    y0, x0 = coords.min(axis=0)
    y1, x1 = coords.max(axis=0) + 1
    return img[y0:y1, x0:x1]


# Pad to 640640
def pad_to_square(img, size=640):
    h, w = img.shape
    scale = size / max(h, w)
    img = cv2.resize(img, (int(w*scale), int(h*scale)))

    h, w = img.shape
    top = (size - h) // 2
    bottom = size - h - top
    left = (size - w) // 2
    right = size - w - left

    img = cv2.copyMakeBorder(img, top, bottom, left, right,
                             cv2.BORDER_CONSTANT, value=0)

    return img


# Convert grayscale -> RGB
def to_rgb(img):
    return cv2.cvtColor(img, cv2.COLOR_GRAY2RGB)


# Full pipeline
def preprocess_dicom(dcm_path, out_path):
    img = load_dicom(dcm_path)
    img = to_gray(img)
    img = resize_image(img)
    img = minmax_normalize(img)
    img = apply_clahe(img)
    img = denoise(img)
    img = crop_borders(img)
    img = pad_to_square(img)
    img = to_rgb(img)

    cv2.imwrite(out_path, img)


# Process all files
input_dir = r"D:\Cliniscan\train"
output_dir = r"D:\Cliniscan\processed_images_cv2"
os.makedirs(output_dir, exist_ok=True)
for file in tqdm(os.listdir(input_dir)):
    if file.endswith(".dicom"):
        try:
            preprocess_dicom(
                os.path.join(input_dir, file),
                os.path.join(output_dir, file.replace(".dicom", ".png"))
            )
        except Exception as e:
            print("Skipping corrupted file:", file, "Error:", e)
            continue
```

```python
import os
import pandas as pd
import numpy as np
import cv2
import pydicom
from tqdm import tqdm

# Paths
train_csv = r"D:\Cliniscan\train.csv"
dicom_dir  = r"D:\Cliniscan\train"
processed_dir = r"D:\Cliniscan\processed_images_cv2"

# YOLO output folders
yolo_img_train = r"D:\Cliniscan\yolo_dataset\images\train"
yolo_lbl_train = r"D:\Cliniscan\yolo_dataset\labels\train"

os.makedirs(yolo_img_train, exist_ok=True)
os.makedirs(yolo_lbl_train, exist_ok=True)

# Load CSV
df = pd.read_csv(train_csv)

def get_original_shape(dcm_path):
    ds = pydicom.dcmread(dcm_path)
    return ds.pixel_array.shape  # (H, W)

def preprocess_mapping(original_h, original_w):
    """
    Returns the scale factor for the resize(1024),
    then crop  pad mapping to 640x640.
    """
    # Step 1: Resize to 1024x1024
    resize_size = 1024
    scale_h = resize_size / original_h
    scale_w = resize_size / original_w

    # Step 2: After crop, we pad to 640
    final_size = 640
    
    return scale_h, scale_w, final_size


def convert_to_yolo(row, scale_h, scale_w, crop_coords, final_size):
    x_min = row.x_min * scale_w
    y_min = row.y_min * scale_h
    x_max = row.x_max * scale_w
    y_max = row.y_max * scale_h

    # Apply cropping offset
    y0, x0 = crop_coords
    x_min -= x0
    x_max -= x0
    y_min -= y0
    y_max -= y0

    # Clamp to image region
    x_min = max(0, x_min)
    y_min = max(0, y_min)
    x_max = max(0, x_max)
    y_max = max(0, y_max)

    # After padding, final image is 640x640
    w = final_size
    h = final_size

    bw = x_max - x_min
    bh = y_max - y_min
    cx = x_min + bw / 2
    cy = y_min + bh / 2

    if bw <= 0 or bh <= 0:
        return None

    return [
        row.class_id,
        cx / w,
        cy / h,
        bw / w,
        bh / h,
    ]


# Main conversion loop
print("Converting annotations to YOLO format...")
for img_file in tqdm(os.listdir(processed_dir)):

    if not img_file.endswith(".png"):
        continue

    image_id = img_file.replace(".png", "")
    dcm_path = os.path.join(dicom_dir, image_id + ".dicom")

    # Skip corrupted/missing DICOM
    if not os.path.exists(dcm_path):
        continue

    # Original DICOM size
    try:
        orig_h, orig_w = get_original_shape(dcm_path)
    except:
        continue  # corrupted dicom

    scale_h, scale_w, final_size = preprocess_mapping(orig_h, orig_w)

    # Load processed PNG to detect crop margins
    img = cv2.imread(os.path.join(processed_dir, img_file), cv2.IMREAD_GRAYSCALE)
    h, w = img.shape  # should be 640x640

    # Find where original crop started
    # Identify the padded area (pure black)
    non_zero = np.argwhere(img > 5)
    if len(non_zero) == 0:
        continue

    y0, x0 = non_zero.min(axis=0)

    # Filter rows for this image
    rows = df[df.image_id == image_id]

    label_path = os.path.join(yolo_lbl_train, image_id + ".txt")
    f = open(label_path, "w")

    for _, row in rows.iterrows():

        # Skip "No finding"
        if np.isnan(row.x_min):
            continue

        yolo_box = convert_to_yolo(row, scale_h, scale_w, (y0, x0), final_size)

        if yolo_box:
            f.write(
                "{} {:.6f} {:.6f} {:.6f} {:.6f}\n".format(
                    int(yolo_box[0]),
                    yolo_box[1], yolo_box[2],
                    yolo_box[3], yolo_box[4]
                )
            )
    f.close()

    # Copy processed image to YOLO folder
    cv2.imwrite(os.path.join(yolo_img_train, img_file), img)

print("YOLO conversion complete.")
```

```python
import os
import random
import shutil

base = r"D:\Cliniscan\yolo_dataset"

img_train = os.path.join(base, "images", "train")
lbl_train = os.path.join(base, "labels", "train")

img_val = os.path.join(base, "images", "val")
lbl_val = os.path.join(base, "labels", "val")
os.makedirs(img_val, exist_ok=True)
os.makedirs(lbl_val, exist_ok=True)

files = os.listdir(img_train)
random.shuffle(files)

val_count = int(0.1 * len(files))
val_files = files[:val_count]

for f in val_files:
    # move images
    shutil.move(os.path.join(img_train, f), os.path.join(img_val, f))
    
    # move labels
    label_name = f.replace(".png", ".txt")
    shutil.move(os.path.join(lbl_train, label_name), os.path.join(lbl_val, label_name))

print("Validation split complete.")
print("Training images:", len(os.listdir(img_train)))
print("Validation images:", len(os.listdir(img_val)))
```
